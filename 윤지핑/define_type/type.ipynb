```python
with open("./travel_personality_type.html",'r') as html:
    html_content = html.read()
html_content
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
print(html_content)
```

```python
headers_to_split_on = [
    ("h3", "Header 1"),  
    ("h4", "Header 2"),
    #("h3", "Header 3"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits = html_splitter.split_text(html_content)

header_1 = []
for html in html_header_splits:
    if "Header 1" in  html.metadata and "Header 2" in html.metadata:
        header_1.append(html.metadata["Header 1"])
header_1 = list(set(header_1))
header_1_dict = {}


for html in html_header_splits:
    if "Header 1" in html.metadata and "Header 2" not in html.metadata:
        if html.metadata['Header 1'] in header_1:
            header_1_dict[html.metadata["Header 1"]] = html.page_content
```

```python
MBTI = ['ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP', 'INTP', 'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']

mbti_dict = {}


for header in html_header_splits:
    if 'Header 2' in header.metadata:
        for m in MBTI:
            if m in header.metadata['Header 2']:
                mbti_dict[m] = f"{header.metadata['Header 1']} {header.metadata['Header 2']}. " + header_1_dict[header.metadata['Header 1']] + " " + header.page_content
```

```python
mbti_dict
```

```python
## source :  MBTI 성격유형에 따른 관광자 행동의 차이: 20대 젊은 층을 대상으로
# E(P,J), I(P,J)에 대한 특징

E = "다른 사람과 여행하는 것을 좋아하고 즐거움이나 재미있는 시설이 포함된 여행 선호합니다. 기쁨과 재미를 찾아 다니며 행사나 축제에 참여하는 걸 선호합니다."
I = "혼자 여행하는 것을 좋아하고 힐링되거나 릴렉스한 여행을 선호합니다. 명소를 보러가거나, 맛집가는 걸 좋아하고 자연스러운 환경에 있는 걸 선호합니다."

for mbti,content in mbti_dict.items():
    if 'E' in mbti:
        mbti_dict[mbti] = content + " "+ E
    elif 'I' in mbti:
        mbti_dict[mbti] = content + " "+ I
```
