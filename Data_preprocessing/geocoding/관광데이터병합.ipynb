```python
import os
import pandas as pd
import glob
```

```python


def merge_csv_files(root_folder):
    # 모든 CSV 파일의 경로를 저장할 리스트
    all_csv_files = []
    
    # 주어진 폴더와 그 하위 폴더를 순회하며 CSV 파일 찾기
    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.endswith('.csv'):
                all_csv_files.append(os.path.join(root, file))
    
    # 찾은 CSV 파일이 없으면 오류 메시지 출력
    if not all_csv_files:
        print("CSV 파일을 찾을 수 없습니다.")
        return None
    
    # 모든 CSV 파일을 데이터프레임으로 읽어 리스트에 저장
    df_list = [pd.read_csv(file) for file in all_csv_files]
    
    # 모든 데이터프레임을 하나로 합치기
    merged_df = pd.concat(df_list, ignore_index=True)
    
    return merged_df

# 사용 예시
root_folder = '생성형AI분야_로우데이터/제주_관광_데이터'  # 여기에 실제 폴더 경로를 입력하세요
result = merge_csv_files(root_folder)

if result is not None:
    print(f"병합된 데이터프레임의 shape: {result.shape}")
    # 결과를 새로운 CSV 파일로 저장
    result.to_csv('merged_output.csv', index=False)
    print("병합된 CSV 파일이 'merged_output.csv'로 저장되었습니다.")
```

```python
def preprocess_address(address):
    # NaN 값 체크
    if pd.isna(address):
        return ''
    
    # 문자열로 변환
    address = str(address)
    
    # '제주특별자치도' 제거
    address = address.replace('제주특별자치도 ', '')
    
    # # '제주' 제거 (일부 주소에서 '제주특별자치도' 대신 '제주'로 시작하는 경우)
    # address = address.replace('제주 ', '')
    
    # 괄호와 그 안의 내용 제거
    address = re.sub(r'\([^)]*\)', '', address)
    
    # 주소를 공백으로 분리
    parts = address.split()
    
    # 뒤에서부터 검사하여 조건에 맞는 마지막 단어 찾기
    for i in range(len(parts) - 1, -1, -1):
        if re.match(r'^\d{2,}$', parts[i]) or '-' in parts[i]:
            # 조건에 맞는 단어까지 슬라이스
            cleaned_address = ' '.join(parts[:i+1])
            break
    else:
        # 조건에 맞는 단어를 찾지 못한 경우, 원래 주소 사용
        cleaned_address = address
    
    # 연속된 공백을 하나의 공백으로 대체
    cleaned_address = re.sub(r'\s+', ' ', cleaned_address).strip()
    
    return cleaned_address
```

```python
orgad['ADR'] = orgad['ADDR'].apply(preprocess_address)
```
